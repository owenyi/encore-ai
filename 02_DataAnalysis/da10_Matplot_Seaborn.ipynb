### 직선 혹은 꺾은선을 그래프 그리기

### 겹쳐진 그래프 그리기

```python
import numpy as np
```

```python
t = np.arange(0, 5, 0.2) # 0에서 5미만...0.2간격으로 arange
t

plt.plot(t, t, 'b--')
plt.show()
```

```python
plt.plot(t, t, 'b--', t, t**2, 'gs') # green square
plt.show()
```

```python
plt.plot(t, t, 'b--', t, t**3, 'r^') # red triangle
plt.show()
```

```python
plt.plot(t, t, 'b--', t, t**2, 'gs', t, t**3, 'r^')
plt.show()
# 여러 개가 되면 라벨링이 필요함
```

```python
# 겹쳐진 그래프일 때는 각 그래프에 대한 정보를 표시하는 게 좋다

plt.plot(t, t, 'b--', label='A Group')
plt.plot(t, t**2, 'gs', label='B Group')
plt.plot(t, t**3, 'r^', label='C Group')
plt.legend() # 라벨링 표시
plt.title('Wrapped Graph')
plt.xlabel('Year')
plt.ylabel('Growth')
plt.grid() # 모눈종이
plt.show()
```

### 파형모양의 그래프

```python
t = np.arange(0, 12, 0.1)

plt.plot(t, np.sin(t), 'r')
plt.plot(t, np.cos(t), 'g')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph Test2')
plt.show()
```

```python
t = np.arange(0, 12, 0.1)

# 기본으로 존재하는 figure가 아닌 새로운 figure를 생성하기 위해서는 plt.figure()를 사용한다.
plt.figure(figsize=(8, 6))
plt.plot(t, np.sin(t), 'r', lw=13) # lw : line width
plt.plot(t, np.cos(t), 'g')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph Test2')
plt.show()
```
