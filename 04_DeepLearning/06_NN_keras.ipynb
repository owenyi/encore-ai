### Keras를 이용한 NN
    가장 단순한 모델
    입력 하나 - 출력 하나
    레이어 없음
    
    케라스 모델 설꼐 진행순서
    1) 데이터 셋 생성
    2) 시퀀스 모델 생성
    3) compile() 사용해서 loss, optim 선정의
       학습을 어떻게 할지 정의
    4) fit() 사용해서 학습 진행
    5) predict() 예측하기
    6) evaluate() 평가하기
    7) accuracy() 정확도 측정

```python
import numpy as np
import keras
```

```python
# 1) 데이터 셋 생성
x = np.array([0, 1, 2, 3, 4])
y = 2*x + 1
print(x)
print(y)
```

```python
# 2) 모델 생성 Sequential() -> 여기다가 model...layer를 하나씩 추가하는 방법(add)
model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_shape=(1,))) # Dense() == FCN
```

```python
# 3) compile() 사용해서 어떻게 학습할지 정의
model.compile(optimizer=keras.optimizers.SGD(0.01), # lr = 0.01
              loss='mse', 
              metrics=['accuracy']) # 평가 기준
```

```python
# 4) fit()
model.fit(x[:3], y[:3], 
          epochs=1000, 
          verbose=2) # verbose : {0: silent, 1: progress bar, 2: one line per epoch}
```

```python
# 5) predict()
print("Targets :", y[3:])
print("Predicts :", model.predict(x[3:]).flatten()) # flatten() : 2차원 -> 1차원
```

```python
# 6) evaluate()
eval_value = model.evaluate(x[3:], y[3:])
eval_value # [loss, accuracy]
```

```python
loss_function = model.evaluate(x[3:], y[3:])
print("loss : {}, accuracy : {}".format(*loss_function))
```
